```python
model = keras.Sequential([
    keras.layers.Dense(10,input_shape = (784,),activation = "sigmoid")
])
```

```python
model.compile(optimizer = "adam", 
              loss = "SparseCategoricalCrossentropy",
              metrics = ["Accuracy"])
```

```python
X_test_flatten = X_test.reshape(len(X_test),28*28)
```
